#Simple Moving Average(for smoothing of the graph)

Disadvantage - Similar importance to all the data

```python
df_1['Close:30 days rolling']=df['Close'].rolling(30).mean()
```

```python
df_1[['Close','Close:30 days rolling']].plot(figsize=(12,5))
```

Cummulative Moving Average(CMA)

```python
#Expanding
df_1['Close'].expanding().mean().plot(figsize=(12,5))
```

Exponential Moving Average (EMA)

```python
df_1['EMA_0.1']=df['Close'].ewm(alpha=0.1,adjust=False).mean() #alpha is a smoothing factor
df_1['EMA_0.3']=df['Close'].ewm(alpha=0.3,adjust=False).mean()
```

```python
df_1[['Close','EMA_0.1','EMA_0.3']].plot(figsize=(12,5))
```

```python
from statsmodels.tsa.stattools import adfuller
```

```python
def adf_test(series):
    result=adfuller(series)
    print('ADF Statistics: {}'.format(result[0]))
    print('p-value: {}'.format(result[1]))
```

```python
adf_test(df_1.Close)
```

```python
df_1['Close_fist_difference']=df_1.Close-df_1.Close.shift(1)
```

```python
adf_test(df_1['Close_fist_difference'].dropna())
```
